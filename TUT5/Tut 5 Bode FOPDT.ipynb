```python
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
K = 0.6423
θ = 17.397
τ = 150.312  
```

```python
ω = numpy.logspace(-5, 0, 100)
s = ω*1j

def bode(G):
    fig, [ax_mag, ax_phase] = plt.subplots(2, 1)
    ax_mag.loglog(ω, numpy.abs(G))
    ax_phase.semilogx(ω, numpy.unwrap(numpy.angle(G)))
    
G1 = K*numpy.exp(-θ*s)/(τ*s + 1)
bode(G1)
```
